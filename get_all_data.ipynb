```python
def get_numbers_from_filename(filename):
    return re.search(r'\d+', filename).group(0)
```

## Get whole trager name

```python
def get_whole_traeger_name(df):
    df['T_Name1'] = df['T_Name1'].astype(str) + ', ' + \
                       df['T_Name2'].astype(str) + ', ' + \
                       df['T_Name3'].astype(str) + ', ' + \
                       df['T_Name4'].astype(str)
    return df
```

## Add more columns

```python
def label_art(row):
    if row == 1 :
        return "Uniklinik"
    if row == 2 :
        return "Plankrankenhaus"
    if row == 3 :
        return "mit Versorgungsvertrag"
    if row == 4 :
        return "ohne Versorgungsvertrag"
    if row == 5 :
        return "Bundeswehrkrankenhaus"
```

```python
def clean_betreiber(x):
    if 'HELIOS' in x:
        return 'HELIOS'
    if 'ASKLEPIOS' in x:
        return 'ASKLEPIOS'
    if 'DRK' in x:
        return 'DRK'
    if 'PARACELSUS' in x:
        return 'PARACELSUS'
    if 'MEDECO' in x:
        return 'MEDECO'
    if 'RHÖN' in x:
        return 'RHÖN'
    if 'RHOEN' in x:
        return 'RHÖN'
    if 'SANA' in x:
        return 'SANA'
    if 'AHG' in x:
        return 'AHG'
    if 'AMEOS' in x:
        return 'AMEOS'
    if 'CAPIO' in x:
        return 'CAPIO'
    if 'CELENUS' in x:
        return 'CELENUS'
    if 'BECKER' in x:
        return 'DR. BECKER'
    if 'KMH' in x:
        return 'KMH'
    if 'MEDICLIN' in x:
        return 'MEDICLIN'
    if 'SCHÖN' in x:
        return 'SCHÖN'
    if 'SCHOEN' in x:
        return 'SCHÖN'
    if 'SRH' in x:
        return 'SRH'
    
    #TODO Check for other betreiber
    else:
        return x
```

```python
def check_traeger(x):
    if 1 <= x <= 4:
       return "oeffentlich"
    elif x == 5:
       return "privat"
    else:
       return "keine Angabe"
```

```python
def add_more_columns(df):
    df.Traeger = df.Traeger.fillna(0)
    df.Traeger = df.Traeger.astype('int')
    df["private_traeger"]= df.Traeger.apply(check_traeger)
    # np.where(cond1, exp1, np.where(cond2, exp2, ...))
    # df.Traeger.apply(lambda x: "oeffentlich" if 1 <= x <= 4 elif x == 5 "privat" else "keine Angabe")

    df["art_text"]= df.Art.apply(lambda x: label_art(x))
    df["betreiber"] = df.T_Name1.str.strip().str.upper()
    df.betreiber = df.betreiber.astype('str', errors='ignore')
    df.betreiber = df.betreiber.apply(clean_betreiber)
    df.Betten_Ins = df.Betten_Ins.astype('int', errors='ignore')
    return df
```

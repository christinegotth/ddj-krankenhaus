## Add more columns

```python
def label_art(row):
    if row == 1 :
        return "Uniklinik"
    if row == 2 :
        return "Plankrankenhaus"
    if row == 3 :
        return "mit Versorgungsvertrag"
    if row == 4 :
        return "ohne Versorgungsvertrag"
    if row == 5 :
        return "Bundeswehrkrankenhaus"
```

```python
def clean_betreiber(x):
    if 'HELIOS' in x:
        return 'HELIOS'
    if 'ASKLEPIOS' in x:
        return 'ASKLEPIOS'
    if 'DRK' in x:
        return 'DRK'
    #TODO Check for other betreiber
    else:
        return x
```

```python
def check_traeger(x):
    if 1 <= x <= 4:
       return "oeffentlich"
    elif x == 5:
       return "privat"
    else:
       return "keine Angabe"
```

```python
def add_more_columns(df):
    df.Traeger = df.Traeger.fillna(0)
    df.Traeger = df.Traeger.astype('int')
    df["private_traeger"]= df.Traeger.apply(check_traeger)
    # np.where(cond1, exp1, np.where(cond2, exp2, ...))
    # df.Traeger.apply(lambda x: "oeffentlich" if 1 <= x <= 4 elif x == 5 "privat" else "keine Angabe")

    df["art_text"]= df.Art.apply(lambda x: label_art(x))
    df["betreiber"] = df.T_Name1.str.strip().str.upper()
    df.betreiber = df.betreiber.astype('str', errors='ignore')
    df.betreiber = df.betreiber.apply(clean_betreiber)
    df.Betten_Ins = df.Betten_Ins.astype('int', errors='ignore')
    return df
```

```python
df_1992 = add_more_columns(df_1992)
```

## Wieviele Kliniken je Träger gibt es?

```python
df.describe()
```

```python
df.groupby("Traeger").count()
```

```python
df["private_traeger"]= df.Traeger.apply(lambda x: "oeffentlich" if x <= 4 else "privat")
```

```python
def label_art(row):
    if row == 1 :
        return "Uniklinik"
    if row == 2 :
        return "Plankrankenhaus"
    if row == 3 :
        return "mit Versorgungsvertrag"
    if row == 4 :
        return "ohne Versorgungsvertrag"
    if row == 5 :
        return "Bundeswehrkrankenhaus"
```

```python
df["art_text"]= df.Art.apply(lambda x: label_art(x))
```

```python
df.sample(50)
```

```python
df.groupby("private_traeger").count()
```

## Wieviele Unikliniken sind privat

```python
privat = df[df.private_traeger == "privat"].groupby("art_text").count()
privat = privat.reset_index()
privat
```

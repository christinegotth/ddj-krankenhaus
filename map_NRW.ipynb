```python
import pandas as pd
import folium
from folium.plugins import FastMarkerCluster

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 1862)
```

```python
df = pd.read_pickle("./Points_2015.pkl")
df = df.dropna()
```

```python
mapNRW = folium.Map(
    location=[51.452441, 7.019965],
    tiles='cartodbpositron',
)
```

```python
df.apply(lambda row:folium.Marker(location=[row["latitude"], row["longitude"]]).add_to(mapNRW), axis=1)
```

```python
mapNRW.save("mapNRW.html")
```

## Wieviele private und öffentliche Kliniken gibt es (nach Jahren)

```python
df = pd.read_pickle("./Krankenhäuser.pkl")
```

```python
df_privat = df['private_traeger'].groupby(df['year']).value_counts()
df_privat
```

```python
df_privat.plot(kind="bar")
```
